                if p-value > 0.05 --> accept HO
                   since p-value < 0.05, reject the null hypothesis.
                   there is a statically significant difference between etest_p and mba_p at the 5% signifinacance. 

# Q8)Test the similarity between the degree_t(Sci&Tech) and specialisa-tion(Mkt&HR) with respect to salary at significance level of 5%.(Make)decision using Hypothesis Testing

```python
"""
Notes:
    Hypothesis
        1. Null Hypothesis (H0) : There is no significant difference between the means of etest_p and mba_p.
        2. Alternative Hypothesis (H1) : There is a significant difference between the means of etest_p and mba_p
"""
```

```python
from scipy.stats import ttest_ind 
group1 = dataset[dataset['degree_t']=="Sci&Tech"]["salary"]
group2 = dataset[dataset["specialisation"]=="Mkt&HR"]["salary"]
```

```python
stats.f_oneway(group1,group2)
```

```python
t_statistic, p_value = ttest_ind(group1, group2, equal_var = False)
```

```python
print("T_statistic", t_statistic)
print("P_Value", p_value)
```

# Explanation

        Decision Rule:
                if p-value < 0.05 ---> Reject H0
                if p-value > 0.05 ---> Accept H0
                Since the p-value (0.0111) is less than 0.05, we reject the null hypothesis. Hence there is a significant difference 
                between the average salaries of Sci&Tech and Mkt&HR students.

# Q 10)What is the probability Density Function of the salary range from 700000 to 900000?

```python
import seaborn as sns
x = dataset['salary']
ax = sns.displot(dataset['salary'], kde=True)
```

```python
def get_pdf_probablity(dataset,startrange,endrange):
    from matplotlib import pyplot
    import seaborn as sns 
    from scipy.stats import norm
    ax = sns.displot(dataset, kde = True, color='Green')
    pyplot.axvline(startrange, color = 'Red')
    pyplot.axvline(endrange, color = 'Red')

    sample = dataset

    sample_mean = sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' %(sample_mean, sample_std))

    dist = norm(sample_mean, sample_std)

    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]
    prob = sum(probabilities)
    print("the area between range ({}, {}):{}". format(startrange, endrange, sum(probabilities)))

    return prob
```

```python
get_pdf_probablity(dataset['salary'],700000,900000)
```

```python
salary1 = dataset[dataset["salary"]!=0]["salary"]
```

```python
mean1 = salary1.mean()
std1 = salary1.std()
print(mean1, std1)
```

```python
from scipy.stats import norm 
probability = norm.cdf(900000,mean1,std1)-norm.cdf(700000,mean1,std1)
print("Probabity:", probability)
```

# Explaination
        - this is calculate the probability of observing salaries between 7,00,000 and 9,00,000  assuming a normal distribution
        - the probability of salary falling between 7,00,000 and 9,00,000 is approximately 0.00000597, indicating that such salaries as very rare.

# 11)Test the similarity between the degree_t(Sci&Tech)with respect to etest_p and mba_p at significance level of 5%.(Make decision using Hy-pothesis Testing)

```python
from scipy.stats import ttest_ind 
sci = dataset[dataset["degree_t"]=="Sci&Tech"]
etest = sci["etest_p"]
mba = sci["mba_p"]
t_stat, p_value = ttest_ind(etest,mba)
print("T-Statistics", t_stat)
print("P-value", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
    
```

# Explaination:
        A t-test is used to determine whether the means differ significantly.
        Based on the obtained p-value, the hypothesis decision was made at the 5% siggnificance level.
